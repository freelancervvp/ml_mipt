```python
import pandas as pd
import numpy as np
from sklearn.cluster import MeanShift
```

```python
checkins = pd.read_csv('fsq/umn_foursquare_datasets/checkins.dat',sep='|', header=0, low_memory=False)
```

```python
checkins.shape
```

```python
checkins.head()
```

```python
checkins.columns.tolist()
```

```python
checkins.columns = ['id','user_id','venue_id','latitude','longitude','created_at']
```

```python
checkins_mapped = checkins[(checkins['latitude'] != '                   ') & (checkins['longitude'] != '                   ')]
```

```python
checkins_mapped.shape
```

```python
checkins_mapped.reset_index(drop=True)
```

```python
checkins_mapped_subset = checkins_mapped[['latitude', 'longitude']][:100001]
checkins_mapped_subset.reset_index(drop=True)
```

```python
ms = MeanShift(bandwidth=0.1, bin_seeding=True)
ms.fit(checkins_mapped_subset)
labels = ms.labels_
cluster_centers = ms.cluster_centers_

labels_unique = np.unique(labels)
n_clusters_ = len(labels_unique)

print("number of estimated clusters : %d" % n_clusters_)
```

```python
np.set_printoptions(threshold=np.nan)
labels_count = np.bincount(labels)
print labels_count
```

```python
labels_and_count = pd.concat([pd.DataFrame(labels_unique), pd.DataFrame(labels_count)], axis=1)
labels_and_count.columns = ['labels_unique', 'labels_count']
labels_and_count
```

```python
labels_and_count_large = labels_and_count[labels_and_count['labels_count'] > 15]
labels_and_count_large
```

```python
cluster_centers[labels_and_count_large['labels_unique']].shape
```

```python
cluster_centers_large = cluster_centers[labels_and_count_large['labels_unique']]
cluster_centers_large
```

```python
cluster_centers_large_pd = pd.DataFrame(cluster_centers_large)
cluster_centers_large_pd[2] = np.nan
cluster_centers_large_pd[3] = np.nan
cluster_centers_large_pd[3] = cluster_centers_large_pd[3].apply(str)
cluster_centers_large_pd.columns = ['lat', 'lon', 'dist', 'city']
print cluster_centers_large_pd.dtypes
cluster_centers_large_pd
```

```python
cluster_centers_large_pd.loc[0, 'city'] = 'and'
cluster_centers_large_pd
```

```python
offices = np.array([[33.751277, -118.188740, 'LA'], [25.867736, -80.324116, 'Miami'], [51.503016, -0.075479, 'London'], 
                   [52.378894, 4.885084, 'Amsterdam'], [39.366487, 117.036146, 'Beijing'], [-33.868457, 151.205134, 'Sydney']])
offices
```

```python

for i in range(len(cluster_centers_large_pd)):
    dist_min = 999999999
    closest_office = ''
    for j in range(len(offices)):
        dist = np.sqrt(np.power(cluster_centers_large_pd.ix[i][0] - np.float(offices[j][0]), 2) + 
                       np.power(cluster_centers_large_pd.ix[i][1] - np.float(offices[j][1]), 2))
        if dist < dist_min:
            dist_min = dist
            closest_office = offices[j][2]
    #cluster_centers_large_pd.ix[i]['dist'] = dist_min
    #cluster_centers_large_pd.ix[i]['city'] = closest_office
    cluster_centers_large_pd.loc[i, 'dist'] = dist_min
    cluster_centers_large_pd.loc[i, 'city'] = closest_office
```
