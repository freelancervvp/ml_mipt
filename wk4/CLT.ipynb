```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

# Распределение Эрланга

Сгенерируем выборку из случ величины распределенной по закону Эрланга с параметром а = 2

```python
a = 2.0

# зададим нормально распределенную случайную величину
erlang = sts.erlang(a)

# сгенерируем выборку из 1000 значений
erlang_sample = erlang.rvs(size = 1000)
```

Построим график функции распределения

```python
x = np.linspace(0,10,100)
cdf = erlang.cdf(x) # функция может принимать и вектор (x)
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

Построим график теоретич плотности распределения и гистограмму выборки

```python
plt.hist(erlang_sample, bins = 50, normed = True)
x = np.linspace(0,10,100)
pdf = erlang.pdf(x)
plt.plot(x, pdf,color = 'red')
plt.ylabel('$P(X=x)$ | $Hist$')
plt.xlabel('$x$')
```

Сгенерируем по 1000 выборок объема n=5,10,50

```python
erlang_sample_5 = []
erlang_sample_10 = []
erlang_sample_50 = []
for i in range(1000):
    erlang_sample_5.append(erlang.rvs(size = 5))
    erlang_sample_10.append(erlang.rvs(size = 10))
    erlang_sample_50.append(erlang.rvs(size = 50))
erlang_sample_5_means = np.mean(erlang_sample_5, axis = 1)
erlang_sample_10_means = np.mean(erlang_sample_10, axis = 1)
erlang_sample_50_means = np.mean(erlang_sample_50, axis = 1)

plt.figure(1, figsize = (20,10))
plt.subplot(131)
plt.hist(erlang_sample_5_means, bins = 50, normed = True, label='n=5', color = 'red')
plt.ylabel('$Hist$')
plt.xlabel('$x$')
plt.legend()

plt.subplot(132)
plt.hist(erlang_sample_10_means, bins = 50, normed = True, label='n=10', color = 'green')
plt.ylabel('$Hist$')
plt.xlabel('$x$')
plt.legend()

plt.subplot(133)
plt.hist(erlang_sample_50_means, bins = 50, normed = True, label='n=50', color = 'blue')
plt.ylabel('$Hist$')
plt.xlabel('$x$')
plt.legend()
```

Рассчитаем параметры mu/sigma для нормальных распределений, описывающих выборочное среднее наших выборок

```python
mu = a # a=2
sigma_5 = math.sqrt(a / 5)
sigma_10 = math.sqrt(a / 10)
sigma_50 = math.sqrt(a / 50)

# зададим нормальные распределения
norm_5 = sts.norm(loc=mu, scale=sigma_5)
norm_10 = sts.norm(loc=mu, scale=sigma_10)
norm_50 = sts.norm(loc=mu, scale=sigma_50)

# построим их поверх наших гистограмм
plt.figure(1, figsize = (20,10))
x = np.linspace(0,5,100)

plt.subplot(131)
plt.hist(erlang_sample_5_means, bins = 50, normed = True, label='n=5', color = 'red')
plt.plot(x, norm_5.pdf(x), 'k-', lw=2, label='pdf')
plt.ylabel('$P(X=x)$ | $Hist$')
plt.xlabel('$x$')
plt.legend()

plt.subplot(132)
plt.hist(erlang_sample_10_means, bins = 50, normed = True, label='n=10', color = 'green')
plt.plot(x, norm_10.pdf(x), 'k-', lw=2, label='pdf')
plt.ylabel('$P(X=x)$ | $Hist$')
plt.xlabel('$x$')
plt.legend()

plt.subplot(133)
plt.hist(erlang_sample_50_means, bins = 50, normed = True, label='n=50', color = 'blue')
plt.plot(x, norm_50.pdf(x), 'k-', lw=2, label='pdf')
plt.ylabel('$P(X=x)$ | $Hist$')
plt.xlabel('$x$')
plt.legend()
```
