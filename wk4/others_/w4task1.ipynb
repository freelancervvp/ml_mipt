### Расчет

```python
#Подключим нужные библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
#Построим случайную выборку объема 1000 для распределение парето (k=3)
pareto_rv=sts.pareto(3,0)
smp=pareto_rv.rvs(1000)
```

```python
#Построим гистограмму плотности распределения и нарисуем теоретическую плотность распределения
plt.hist(smp,normed=True,label='Hist PDF')
x = np.linspace(1,8,100)
pdf=pareto_rv.pdf(x)
plt.plot(x, pdf, 'r-', label='Theoretical PDF')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.xlim(1,8)
plt.show()
```

```python
#Сгенерируем 1000 выборок размера 5,10,50,100
pareto_rv=sts.pareto(3)
smp_n5 = [pareto_rv.rvs(5) for x in range(0, 1000)]
smp_n10 = [pareto_rv.rvs(10) for x in range(0, 1000)]
smp_n50 = [pareto_rv.rvs(50) for x in range(0, 1000)]
smp_n100 = [pareto_rv.rvs(100) for x in range(0, 1000)]
```

```python
#Расчитаем выборочные средние для каждого набора выборок
smp_n5_mean=[np.mean(smp_n5[x]) for x in range(0, 1000) ]
smp_n10_mean=[np.mean(smp_n10[x]) for x in range(0, 1000) ]
smp_n50_mean=[np.mean(smp_n50[x]) for x in range(0, 1000) ]
smp_n100_mean=[np.mean(smp_n100[x]) for x in range(0, 1000) ]
```

```python
#Построим гистограмму выборочных средний выборок, где n=5, и теоретическую плотность cоответсвующего нормального распределения
norm_rv = sts.norm(pareto_rv.mean(), np.sqrt(pareto_rv.var()/5))
x = np.linspace(0,4,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, 'r-', label='Theoretical PDF')

plt.hist(smp_n5_mean,20,normed=True,label='n=5')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
#Построим гистограмму выборочных средний выборок, где n=10, и теоретическую плотность нормального распределения
norm_rv = sts.norm(pareto_rv.mean(), np.sqrt(pareto_rv.var()/10))
x = np.linspace(0,4,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, 'r-', label='Theoretical PDF')

plt.hist(smp_n10_mean,20,normed=True,label='n=10')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
#Построим гистограмму выборочных средний выборок, где n=50, и теоретическую плотность нормального распределения
norm_rv = sts.norm(pareto_rv.mean(), np.sqrt(pareto_rv.var()/50))
x = np.linspace(0,4,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, 'r-', label='Theoretical PDF')

plt.hist(smp_n50_mean,20,normed=True,label='n=50')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
#Построим гистограмму выборочных средний выборок, где n=100
norm_rv = sts.norm(pareto_rv.mean(), np.sqrt(pareto_rv.var()/100))
x = np.linspace(0,4,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, 'r-', label='Theoretical PDF')

plt.hist(smp_n100_mean,20,normed=True,label='n=100')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

### Выводы
#### Разница между полученными распределениями при различных значениях n
При увеличении размера выборки:
1. Занчения плотнее группируются вокруг средней (для распределения Парето с праметроми 3,0 - это 1,5)
2. Т.к. для данного распределение х >= 1, то с увеличени n выборочные среднии удаляются от единицы

#### Как меняется точность аппроксимации распределения выборочных средних нормальным с ростом n?
С ростом n точность аппроксимации увеличивается и теоретическая плотность распределения приближается к распределению выборочных средних.
