# Непрерывное распределение Фиска

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Сгенерируем выборку объёма 1000 из непрерывного распределения Фиска ($k = 3.1$)

```python
k = 3.1
fisk_rv = sts.fisk(k)
sample = fisk_rv.rvs(1000)
```

Гистограмма выборки и теоретическая плотность распределения:

```python
plt.figure(figsize=(15,7))
x = np.linspace(-2,10,100)
pdf = fisk_rv.pdf(x)
plt.plot(x, pdf, label='Theoretical pdf', alpha=1)
plt.hist(sample, bins=150, normed=True, label='Histogram', alpha=0.6)
plt.legend()
```

```python
#Оценим распределение выборочного среднего случайной величины при разных объёмах выборок.
#Сгенерируем 1000 выборок объёмов 5, 15, 30, 50 и построем гистограммы распределений их выборочных средних (Xn).
#Поверх каждой гистограммы нарисуем плотность соответствующего нормального распределения.

mean, var = sts.fisk.stats(k, moments='mv') # Математическое ожидание и дисперися случайной величины
Xn = np.zeros(1000)
for n in [5, 15, 30, 50]:
    for m in range(1000):
        sample = fisk_rv.rvs(n)
        Xn[m] = np.mean(sample)
    #нормальное распределение (c соответствующими параметрами), которым приближается распределение выборочных средних
    norm_rv = sts.norm(loc= mean, scale=np.sqrt(var/n)) 
    x = np.linspace(-1,5,1000)
    pdf = norm_rv.pdf(x)
    plt.figure(figsize=(15,7))
    plt.plot(x, pdf, label='Theoretical pdf', alpha=1)
    plt.hist(Xn, bins=100, normed=True ,label='Histogram', alpha=0.5)
    plt.legend()
    plt.title("$n = %s$"% n)
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    plt.axis('equal')
    plt.show()    
```

По полученным гистограммам и графикам плотностей распределений можно сделать вывод,что с ростом n точность аппроксимации распределения выборочных средних увеличиватся, распределение Фиска точнее описывается нормальным.
Хоть и исходное распределение немного скошено, центральная предельная теорема работает.
